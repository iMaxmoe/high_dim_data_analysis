```python
import pandas as pd
import numpy as np

stocks = pd.read_csv("sample_stocks.csv")
stocks.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6.4))
plt.scatter(stocks.iloc[:,0],stocks.iloc[:,1])
plt.xlabel('Returns')
plt.ylabel('Dividendyield')
plt.title('Scatter Plot')
plt.show()
```

```python
# Determine the number of clusters
from sklearn.cluster import KMeans
wss = []
for i in range(2,11):
    km = KMeans(n_clusters=i, random_state=0)
    km.fit_predict(stocks)
    wss.append(km.inertia_)
plt.xlabel("# of Clusters")
plt.ylabel("SSE")
plt.plot(wss,'-o')
plt.show()
```

```python
km =  KMeans(n_clusters=3, random_state=0)
cluster = km.fit_predict(stocks)
color_labels = {0:'r',1:'g',2:'b'}
label_color = [color_labels[i] for i in cluster]
plt.scatter(stocks.iloc[:,0], stocks.iloc[:,1],c=label_color,alpha=0.9)
plt.xlabel('Returns')
plt.ylabel('Dividendyield')
plt.title('Scatter Plot')
plt.grid()
plt.show()
```

### Hierarchical Clutsering

```python
import scipy.cluster.hierarchy as hr

plt.figure(figsize=(14, 7))  
plt.title("Dendogram")  
plt.ylabel("Distance")
dend = hr.dendrogram(hr.linkage(stocks, method='complete'))  
```
