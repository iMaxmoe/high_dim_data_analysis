```python
import numpy as np
import pandas as pd

df = pd.read_csv("Mental_Health_survey.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
policies_list = ["benefits", "care_options", "wellness_program", "seek_help", "anonymity", "leave"]
policies_on_MH = df[policies_list]
for i in range(5):
    y = policies_on_MH.iloc[:,i]
    y = enc.fit_transform(y) - 1
    policies_on_MH.iloc[:,i] = y
    
y = policies_on_MH.loc[:,"leave"]
y = enc.fit_transform(y) - 2
policies_on_MH.loc[:,"leave"] = y
```

```python
openness_list = ["mental_health_consequence", "phys_health_consequence", "coworkers",
                        "supervisor", "mental_health_interview", "phys_health_interview",
                        "mental_vs_physical", "obs_consequence"]
openness_of_MH = df[openness_list]
for i in range(7):
    y = openness_of_MH.iloc[:,i]
    y = enc.fit_transform(y) - 1
    openness_of_MH.iloc[:,i] = y
    
y = openness_of_MH.iloc[:,-1]
y = enc.fit_transform(y)
openness_of_MH.iloc[:,-1] = y
```

```python
from sklearn.cross_decomposition import CCA

#policies_on_MH = policies_on_MH.reset_index()
cca = CCA()
cca.fit_transform(policies_on_MH,openness_of_MH)
```
